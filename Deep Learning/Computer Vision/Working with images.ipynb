```python
def filtering(img, f = 2):
    (rows, columns, channels) = img.shape
    
    stride = 2
    
    n_rows = int(1 + (rows - f)/stride)
    n_cols = int(1 + (columns - f)/stride)
    n_channels = channels
    
    # output channel
    n_img = np.zeros((n_rows, n_cols, n_channels))
    
    for h in range(n_rows):
        for w in range(n_cols):
            for c in range(n_channels):
                vert_start = stride * h
                vert_end = vert_start + f
                hor_start = stride * w
                hor_end = hor_start + f
                
                n_slice = img[vert_start:vert_end, hor_start:hor_end]
                
                n_img[h, w, c] = np.mean(n_slice, dtype=int)
    return n_img
```

```python
A = filtering(img)
plt.imshow(A)
```
