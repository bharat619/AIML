```python
import numpy as np
import pandas as pd
import keras
from keras.datasets import mnist
```

```python
(x_train,y_train), (x_test, y_test) = mnist.load_data()
```

```python
X = np.zeros((x_train.shape[0], 784))
```

```python
x_train.shape
```

```python
X.shape
```

```python
for i in range(x_train.shape[0]):
    X[i] = x_train[i].flatten()
```

```python
X = pd.DataFrame(X)
```

```python
Y = pd.DataFrame(y_train)
```

```python
Y.head()
```

```python
X = X.sample(frac=0.1, random_state=10)
```

```python
Y = Y.sample(frac=0.1, random_state=10)
```

```python
X.shape
```

```python
Y.shape
```

```python
df = X
```

```python
import time
from sklearn.manifold import TSNE

n_sne = 7000

time_start = time.time()
tsne = TSNE(n_iter=300)
tsne_results = tsne.fit_transform(df.values)

print('t-SNE done. Time elapsed {} seconds'.format(time.time() - time_start))
```

```python
df['label'] = Y
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(12,12))
plt.scatter(tsne_results[:,0], tsne_results[:,1], c=df['label'], cmap=plt.cm.get_cmap('jet',10), alpha=0.8)
plt.colorbar(ticks=range(10))
plt.clim(-0.5,9.5)
```
