## Random Variable
A random variable is a variable that takes on numerical values as a result of random experiment or measurement.
Associates a numerical value with each possible outcome

Things to keep in mind:
1. The random variable
2. Possible outcome the random variable can take

## Discrete Vs Continuous Random Variable
- Discrete random variable has finite number of values or infinite sequence of values (0,1,2,....) and the differences in the outcome are meaningful. 
Example: Die throw can have a possible of 6 values 1,2,3,4,5,6

- Continuous random variable has infinite number of outcomes that cannot be easily counted and the differences between outcome are not meaningful.
Example: With average income difference between $40000.00 and $40000.99 is not meaningful

## Measures of Variability
Consider 2 classes with scores and their mean as
Class 1# 85 95 80 75 90, mean 85
Class 2# 88 79 91 85 82, mean 85

Both class have same mean but data points are different.
Both have the same mean, but class 1 has more variability ( scores are more spread out )
Class 1 has highest score 95 class 2 has 91
Class 1 has lowest score 75 class 2 has 79
Same mean but many variation

How far is the data point from the mean? (Distance)
SD and variance helps to answer these questions

Mean, variation and SD are usefult when comparing a dataset or comparing against a therotical value

### Coefficient of variation
CV = (SD/Mean * 100) %

It is the relative measure of variability
It measures SD relative to the mean ( % of mean related to SD )
CV for class 1 = 9.29%
CV for class 2 = 5.58%


# Binomial Distribution
Unlike the other probability outcome (example die roll which has 6 outcomes ) binomial distribution has only 2 outcomes, either a success or a failure
Binomial experiment has following characteristics
1. The process consists of sequence of n trials
2. Only two exclusive outcomes in each trial -> success/failure
3. Probability of success (p) does not change from trial to trial. Probability of failiure is 1-p and is also fixed from trial to trial.
4. The trials are independent. The outcome of previous trials are independent of next trial.

C(n,x) * p^x * (1-p)^(n-x)
n -> number of trials
x -> numbe4r of success
p -> probability of success in any trial

### Cumulative probability
