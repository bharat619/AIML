```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
x
```

```python
y
```

```python
# Functional method
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,'r')

plt.subplot(1,2,2)
plt.plot(y,'b')
```

```python
# Object oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5, 0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title("Smaller plot")
axes1.set_title("Larger plot")
```

```python
fig,axes =plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title("First")
axes[1].set_title("Second")
plt.tight_layout()
```

```python
# Figure size and DPI
# figsize is tuple with args as width and height
fig = plt.figure(figsize=(8,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(8,2), nrows=2,ncols=1)
axes[0].plot(x,y)

axes[1].plot(y,x)
```

```python
# Adding legend
fig = plt.figure(figsize=(3,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="X Square")

ax.plot(x,x**3, label="X Cube")
ax.legend(loc=0)

# can also specify location
ax.legend(loc=(0.1,0.1))
```

```python
fig = plt.figure()
axes=fig.add_axes([0,0,1,1])
# color can be simple string or rgb hex code
# axes.plot(x,y,'#FF8800')

# line thickness can be increased with linewidth (lw)
# alpha denotes the transparency
# linestyle denotes style of line (ls)
# marker denotes the symbol of the plot x vs y
# markersize denotes size of marker
# markerfacecolor - color inside circle
# markeredgecolor - color of maker edge
# markeredgewidth - thickness of marker edge
axes.plot(x,y, 'orange', linewidth=3, alpha=0.3, linestyle='--', marker="o", markersize=20, markerfacecolor='red',
         markeredgecolor='black',markeredgewidth=3)
```

```python
fig = plt.figure()
axes=fig.add_axes([0,0,1,1])

# to limit x and y values
axes.set_xlim((0,1))
axes.set_ylim((0,2))
axes.plot(x,y)
```
