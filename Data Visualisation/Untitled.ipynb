## Distribution Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Distplot allows us to show univariate set of observations
# univariate- one variable
tips[tips['total_bill'].max() == tips['total_bill']]
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
# jointplot - allows to combine two variants of distplot
sns.jointplot(x='total_bill',y='tip',data=tips)
# type allows to change from scatter to different type of graph
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
# pairplot gives pair wise relationships across all numerical columns
sns.pairplot(tips)
```

```python
# hue - specifies categorical name
sns.pairplot(tips, hue='sex', palette='coolwarm')
# palette specifies color
```

```python
# rugplot
sns.rugplot(tips['total_bill'])

# difference between rugplot and histogram - histogram has bars in the range. rugplot has dashes in the range
```

```python
#kde - kernel density estimation
sns.kdeplot(tips['total_bill'])
```

## Categorical Plot

```python
# concerned about seeing behaviour of a categorical column with a numerical or another categorical column
```

```python
# Bar Plot
# general plot to aggregate the categorical plot based on a function (like mean)
sns.barplot(x='sex',y='total_bill',data=tips)
# gives the mean by default
# change estimator object to have different function
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
#countplot
# same as barplot, but estimator is explicitly counting number of occurances
sns.countplot(x='sex',data=tips)
```

```python
# box and violin plot
# used to show disribution of categorical data
# boxplot - distribution of quantitative data ,facilitates comparision between variables
sns.boxplot(x='day',y='total_bill',data=tips)
# hue will allow us to split by another category
# sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
#violin plot is similar to box plot
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

```python
#stripplot
# will draw scatterplot, one variable is categorical
sns.stripplot(x='day',y='total_bill',data=tips)
# jitter will add some noise, and give bit clear view
# sns.stripplot(x='day',y='total_bill',data=tips, jitter=True,hue='sex')
```

```python
#swarmplot - similar to strip plot, points wont overlap.
# a mixture of violin and strip plot
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

```python
# factor plot
# generalised plotting
sns.factorplot(x='day',y='total_bill',data= tips,kind='bar')
```

## Matrix Plots

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
# for heat map to work properly, data should be in matrix form (index and column name match cell value)
# can be done with pivot_table or correlation
tc= tips.corr()
```

```python
sns.heatmap(tc)
# annotation- annotate numerical value
sns.heatmap(tc, annot=True)
# cmap - color
sns.heatmap(tc, annot=True, cmap='coolwarm')
```

```python
print(flights.head())
fp = flights.pivot_table(index='month',columns='year',values='passengers')
fp.head()
```

```python
sns.heatmap(fp)
sns.heatmap(fp,cmap='magma', linewidths=3)
```

## Grids

```python
# to automate subplots based of features in data
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
# facet grid
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
# for scatterplot
# g.map(sns.scatterplot,'total_bill','tip')
```

## Regression Plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips)

sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex',markers=['o','v'])

# to change the size of the marker
# it uses matplot lib underhood. to increase the size of marker, pass dictionary to scatter_kws
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'], scatter_kws={'s':100})
```

```python
# seperating by column, insetead of color. 
sns.lmplot(x='total_bill',y='tip',data=tips, col='sex')

# adding a time factor
sns.lmplot(x='total_bill',y='tip',data=tips, col='sex', row='time')

sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', col='day', row='time', 
           aspect=0.6, size=8)
```
