```python
import pandas as pd
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

```python
auto = pd.read_csv('Automobile (1).csv')
auto.head()
```

```python
# Different types of visualisation to analyse dataset
# 1. Univariate ( looking at only one variable)
# distplot can be used for a continuous variable
# Discrete and continuous difference: https://www.mathsisfun.com/data/data-discrete-continuous.html
sns.distplot(auto['normalized_losses'])
# sns.distplot(auto['engine_size'])
```

```python
sns.distplot(auto['city_mpg'],kde=False, rug=True)
# rug is the number of observations plotted within the bins
# each line indicate single observation
```

```python
# bivariate distribution
# understand the pattern between 2 variable
# jointplot
sns.jointplot(auto['engine_size'],auto['horsepower'])
# histogram is created for horsepower
# histogram is created for engine_size
```

```python
sns.jointplot(auto['engine_size'],auto['horsepower'], kind='reg')
#hexplot
#its a concentration of hexagons

# kind: {"scatter"|"reg"|"hex","kde"}
```

```python
# multivariate distribution
# visuals between multiple continuous variables
sns.pairplot(auto[['normalized_losses','engine_size','horsepower']])
# diagonal has histogram, since x and y are same for diagonal
```

```python
# Categorical and continuous plot

#jitter creates distribution of dots
sns.stripplot(auto['fuel_type'],auto['horsepower'], jitter=True)
```

```python
# swarm plot
# diff in swarm and strip plot: swarm plot doesnt overlap the points like jitter in stripplot
sns.swarmplot(auto['fuel_type'],auto['horsepower'])
```

```python
# box plot
sns.boxplot(auto['number_of_doors'],auto['horsepower'])
# box top line - max
# box bottom line - min
# box bottom edge - Q1 (25%)
# box middle edge - Q2 (50%)
# box top edge - Q3 (75%)
# distribution are skewed to the right( positive direction)
# the top diamond is the outlayers, that dont follow the rule of 1.5*interquartile range
# interquartile range is diff between Q3 and Q1
# distance between Q3 and the record (diamond) is more than the 1.5 * interquartile distribution, then outlayer is calculated

```

```python
#hue - gives additional information
sns.boxplot(auto['number_of_doors'],auto['horsepower'], hue=auto['fuel_type'])
```

```python
# bar plot - to show the volume (amount of occurances)
# operates on full dataset, and show us arbitrary estimate (uses mean by default)
# when there are multiple observation in each category, bar plot uses bootstrapping
# or aggregating to compute confidence interval around estimated values and plots using error bars
# Error bar is present on the top, which talks about confidence interval values
# Bar plot includes 0's as well

sns.barplot(auto['body_style'],auto['horsepower'])
```

```python
# lines on top are confidence level for the distribution of information

sns.barplot(auto['body_style'],auto['horsepower'], hue=auto['engine_location'])
```

```python
# observe frequency of observation for a set of categorical variable or 
# single categorical variable
# we can use countplot for that
sns.countplot(auto['body_style'])
```

```python
sns.countplot(auto['body_style'], hue=auto['engine_location'])
```

```python
# point plot will show single point which shows highest value of bar plot
# also shows confidence interval (if applicable)
# it connects the values together so that we have a line trend
sns.pointplot(auto['fuel_system'],auto['horsepower'])
```

```python
sns.pointplot(auto['fuel_system'],auto['horsepower'], hue=auto['number_of_doors'])
```

```python
# if there is no line between 2 points then there is missing informatin between 2 points( ex: mfi 1bbl)
```

```python
# factor plot is used to draw multipanel categorical plots
sns.factorplot(x='fuel_type', # x axis
               y='horsepower', # y axis
               hue='number_of_doors',# hue 
               col='engine_location', # create plot for engine_location values
              data=auto,kind='swarm')

# various kind: {point,bar,count,box,violin,strip}
```

```python
# linear regression plot
sns.lmplot(x='horsepower',y='peak_rpm',data=auto, hue='fuel_type')
```
