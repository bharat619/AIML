```python
def get_top_n(prediction, n=10):
    top_n = defaultdict(list)
    for uid,iid,r_val,est, _ in prediction:
        top_n[uid].append((iid, est))
    for iid, est in top_n.items():
        est.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = est[:n]
    return top_n
```
