# Q6. Use popularity based recommender model to make predictions and find recommendations for random list of users with inferences

Since this is a popularity-based recommender model, recommendations remain the same for all users

We predict the products based on the popularity. It is not personalized to particular user

```python
merged_df.head()
```

```python
selected = merged_df[['title','release', 'artist_name']]
```

```python
selected.head()
```

```python
selected['release'] = selected['release'].apply(lambda x: x.split(' '))
```

```python
selected['artist_name'] = selected['artist_name'].apply(lambda x: x.split(' '))
```

```python
selected.head()
```

```python
data = selected.iloc[:10000]
for index, row in data.iterrows():
    row['release'] = ''.join(row['release']).lower()
    row['artist_name'] = ''.join(row['artist_name']).lower()
```

```python
data.head()
```

```python
data['bag_of_words'] = ''
for index, row in data.iterrows():
    row['bag_of_words'] = row['release'] + ' ' + row['artist_name']
```

```python
data.set_index('title', inplace=True)
data.drop([x for x in data.columns if x != 'bag_of_words'], axis=1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count = CountVectorizer()
```

```python
count_matrix = count.fit_transform(data['bag_of_words'])
```

```python
print(count_matrix)
```

```python
indices = pd.Series(data.index)
```

```python
indices[:5]
```

```python
count_matrix.todense()
```

```python
print(count_matrix[0:])
```

```python
count.vocabulary_
```

```python
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
cosine_sim
```

```python
data.head()
```

```python
def recommender(name, cosine_sim = cosine_sim):
    recommended_movies = []
    idx = indices[indices == name].index[0]
    series = pd.Series(cosine_sim[idx]).sort_values(ascending= True)
    top_10_index = np.array(series)[:10]
    for i in top_10_index:
        recommended_movies.append(data.index[i])
    return recommended_movies
```

```python
recommender('The Cove')
```
