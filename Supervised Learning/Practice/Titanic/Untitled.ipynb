```python
avg_age = df['Age'].mean()
age_std = df['Age'].std()
age_null_count = df['Age'].isna().count()
```

```python
age_random_values = np.random.randint(avg_age - age_std, avg_age + age_std, size = age_null_count )
```

```python
df['Age'][np.isnan(df['Age'])] = age_random_values
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df['CategoricalAge'] = pd.cut(df['Age'], 5)
df.head(10)
df[['CategoricalAge', 'Survived']].groupby('CategoricalAge').count()
```

```python
df.head()
```

```python
import re
def get_title(name):
    title = re.search(' ([A-Za-z]+)\.', name)
    if title:
        return title.group(1)
    return ''
```

```python
df['Title'] = df['Name'].apply(get_title)
```

```python
pd.crosstab(df['Title'], df['Sex'])
```

```python
# Lets check the survival on category
df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
df['Title'] = df['Title'].replace('Mile', 'Miss')
df['Title'] = df['Title'].replace('Ms', 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')
```

```python
df[['Title', 'Survived']].groupby('Title').count()
```

```python
# Data cleaning
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
```

```python
df['Title'] = le.fit_transform(df['Title'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df.loc[df['Fare'] <= 7.91, 'Fare'] = 0
df.loc[(df['Fare'] >7.91) &  (df['Fare'] <=14.454), 'Fare'] = 1
df.loc[(df['Fare'] > 14.454) & (df['Fare'] <= 31.0), 'Fare'] = 2
df.loc[df['Fare'] > 31.0, 'Fare'] = 3
```

```python
df['Fare'].astype(int)
df['CategoricalAge'].value_counts()
df.loc[df['Age'] <= 16.0, 'Age'] = 0
df.loc[(df['Age'] > 16.0) & (df['Age'] <= 32.0), 'Age'] = 1
df.loc[(df['Age'] > 32.0) & (df['Age'] <= 48.0), 'Age'] = 2
df.loc[(df['Age'] > 48.0) & (df['Age'] <= 64.0), 'Age'] = 3
df.loc[df['Age'] > 64.0, 'Age'] = 4
```

```python
df = df.drop(['PassengerId', 'Ticket', 'Name', 'Cabin', 'SibSp', 'Parch', 'FamilySize', 'CategoricalFare', 'CategoricalAge'], axis=1)
```

```python
df.head(10)
```

```python
# SVM
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X = np.array(df.drop('Survived', axis=1))
y = np.array(df['Survived'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
cross_val_score(model, X = X, y=y).mean()
```

```python
confusion_matrix(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```
