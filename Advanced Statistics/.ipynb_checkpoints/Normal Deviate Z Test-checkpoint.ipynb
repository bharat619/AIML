**H0**: The difference in the mean between the sample BP column and plpulation mean for BP is a statistical fluctuation <br/>
**H1**: The difference in the mean between the sample BP column and population mean for BP is not a statistical fluctiation


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
%matplotlib inline
```

```python
colnames = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
df = pd.read_csv('diabetes (1).csv', names = colnames)
# populatioin mean
Mu = 71.3
# population standard deviation
Sd = 7.2
```

```python
df.head()
```

```python
mean = np.average(df['BloodPressure'])
std_error = Sd/np.sqrt(df.size)
std_error
```

```python
Z_norm_deviate = (mean - Mu)/std_error
Z_norm_deviate
```

```python
# sf -> survival function
# We are comparing H0 != H1, hence it will be a 2 tailed test
# Multiply by number of tails. We are doing a 2 tailed test
p_value = stats.norm.sf(abs(Z_norm_deviate)) * 2
p_value
stats.t.
```

```python
if p_value > 0.05:
    print("Samples are likely to be drawn from same population")
else:
    print("Samples are not drawn from same sample")
```
